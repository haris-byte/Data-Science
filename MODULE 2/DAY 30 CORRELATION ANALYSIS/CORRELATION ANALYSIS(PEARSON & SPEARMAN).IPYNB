```python
from scipy.stats import pearsonr, spearmanr
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## TASK
### Example: Genshin Weapon Stats Correlation
You have a dataset:
- Weapon_Attack (base ATK stat)
- Crit_Damage (critical hit multiplier)
- Substat_Value (extra substat)

Your task is to:
- Check Pearson correlation between numeric columns.
- Check Spearman correlation (especially if values are ranked or uneven).
- Plot heatmap of correlation matrix.
- Interpret whether strong or weak correlation exists.

## STEP 1:
### SAMPLE DATA FRAME OF 5 WEAPONS


```python
DF = pd.DataFrame({
    'Weapon_Name': ['Amber-Swift', 'Mist-Split', 'Gold-Crush', 'Sky-Wing', 'Dagger-Slash'],
    'Attack': [50, 60, 40, 70, 80], 
    'Crit_Damage': [20, 40, 50, 44, 76],
    'Substat': [10, 19, 17, 21, 24]
})
```

### Correlation

```python
# Pearson between Attack and Crit
pearson_corr, p1 = pearsonr(DF['Attack'], DF['Crit_Damage'])

# Spearman between Attack and Substat
spearman_corr, p2 = spearmanr(DF['Attack'], DF['Substat'])

print(f"Pearson Correlation (Attack vs Crit_Damage): {pearson_corr}, p={p1}")
print(f"Spearman Correlation (Attack vs Substat): {spearman_corr}, p={p2}")

```

### Visualization


```python
sns.heatmap(DF[['Attack','Crit_Damage','Substat']].corr(method='pearson'), annot=True, cmap='coolwarm')
```
